```python
import pandas as pd

chunks = []
chunk_size = 50_000

for chunk in pd.read_json("D:/UNI/Semester_7/Big Data/Project/Electronics_5.json", lines=True, chunksize=chunk_size):
    sample_chunk = chunk.sample(frac=0.01)  
    chunks.append(sample_chunk)

df_sample = pd.concat(chunks)

df_sample.to_json("50_000_sample_dataset.json", orient="records", lines=True)

print("Sample size:", len(df_sample))
```

```python
# Load dataset
df = pd.read_json("../data/raw/50_000_sample_dataset.json", lines=True)
```

```python
# Show info
print(df.info())
print(df.head())
```

```python
# Handle missing values
df = df.fillna({
    "vote": 0,
    "reviewText": "",
    "summary": "",
    "style": {}
})
```

```python
# Convert vote to integer
df["vote"] = df["vote"].astype(str).str.replace(",", "")
df["vote"] = pd.to_numeric(df["vote"], errors="coerce").fillna(0).astype(int)
```

```python
# Convert verified to boolean
df["verified"] = df["verified"].astype(bool)
```

```python
# Remove duplicates
df = df.drop_duplicates(subset=["reviewerID", "asin", "unixReviewTime"])
```

```python
# Ensure rating is float
df["overall"] = df["overall"].astype(float)
```

```python
# Final check
print(df.info())
```

```python
df.to_json("../data/processed/50_000_sample_dataset_processed.json", orient="records", lines=True)
```
