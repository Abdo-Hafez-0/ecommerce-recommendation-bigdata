<h1>Query 6 - $gt : Rating Greater Than 4</h1>

```python
high_rating_reviews = reviews_col.find({"overall": {"$gt": 4}})

for review in high_rating_reviews.limit(5):
    pprint(review)
```

<h1>Query 7 - $gte : Rating Greater Than or Equal to 4</h1>

```python
good_reviews = reviews_col.find({"overall": {"$gte": 4}})

for review in good_reviews.limit(5):
    pprint(review)
```

<h1>Query 8 - $lt : Rating Less Than 3</h1>

```python
low_reviews = reviews_col.find({"overall": {"$lt": 3}})

for review in low_reviews.limit(5):
    pprint(review)
```

<h1>Query 9 - $lte : Rating Less Than or Equal to 2</h1>

```python
very_low_reviews = reviews_col.find({"overall": {"$lte": 2}})

for review in very_low_reviews.limit(5):
    pprint(review)
```

<h1>Query 10 - $ne : Verified Not Equal to True</h1>

```python
unverified_reviews = reviews_col.find({"verified": {"$ne": True}})

for review in unverified_reviews.limit(5):
    pprint(review)
```

<h1>Query 11 - $in : Ratings in [1, 5]</h1>

```python
extreme_reviews = reviews_col.find({"overall": {"$in": [1, 5]}})

for review in extreme_reviews.limit(5):
    pprint(review)
```

<h1>Query 12 - $nin : Ratings not in [1, 2]</h1>

```python
not_low_reviews = reviews_col.find({"overall": {"$nin": [1, 2]}})

for review in not_low_reviews.limit(5):
    pprint(review)
```

<h1>Query 13 - Range Query with gte and lte</h1>

```python
mid_reviews = reviews_col.find({"overall": {"$gte": 3, "$lte": 4}})

for review in mid_reviews.limit(5):
    pprint(review)
```

<h1>Query 14 - $and : Verified and High Rating</h1>

```python

verified_good_reviews = reviews_col.find({
    "$and": [
        {"verified": True},
        {"overall": {"$gte": 4}}
    ]
})

for review in verified_good_reviews.limit(5):
    pprint(review)
```

<h1>Query 15 - $or : Rating 1 or Rating 5</h1>

```python
polarized_reviews = reviews_col.find({
    "$or": [
        {"overall": 1},
        {"overall": 5}
    ]
})

for review in polarized_reviews.limit(5):
    pprint(review)
```

<h1>Query 16 - $not : Not Greater Than 3</h1>

```python
not_high_reviews = reviews_col.find({
    "overall": {
        "$not": {"$gt": 3}
    }
})

for review in not_high_reviews.limit(5):
    pprint(review)
```

<h1>Query 17 - $and with Product and Verified</h1>

<h1>Query 19 - $and with Rating Range and Verified</h1>
